```python
import pandas as pd
from faker import Faker
import numpy as np
import random

# 1. Generación de un dataset simulado para declaraciones juradas patrimoniales
fake = Faker('es_AR')
seed = 42
random.seed(seed)
np.random.seed(seed)
num_funcionarios = 30
data_ddjj = {
    'ID_Funcionario': range(1, num_funcionarios + 1),
    'Nombre_Funcionario': [fake.name() for _ in range(num_funcionarios)],
    'Cargo': [random.choice(['Ministro', 'Secretario', 'Director', 'Asesor']) for _ in range(num_funcionarios)],
    'Organismo': [fake.company() for _ in range(num_funcionarios)],
    'Año_Declaracion': [random.randint(2023, 2024) for _ in range(num_funcionarios)],
    'Bienes_Inmuebles_Valor': np.round(np.random.uniform(0, 1500000, num_funcionarios), 2),
    'Bienes_Muebles_Valor': np.round(np.random.uniform(0, 500000, num_funcionarios), 2),
    'Activos_Financieros_Valor': np.round(np.random.uniform(0, 300000, num_funcionarios), 2),
    'Pasivos_Valor': np.round(np.random.uniform(0, 100000, num_funcionarios), 2),
    'Ingresos_Anuales': np.round(np.random.uniform(50000, 200000, num_funcionarios), 2)
}
df_ddjj = pd.DataFrame(data_ddjj)

# 1. Preparar el DataFrame (Asegurar orden y claves):
df_ddjj = df_ddjj.sort_values(by=['ID_Funcionario', 'Año_Declaracion'])

# 2. Crear una columna "Patrimonio Total":
df_ddjj['Patrimonio_Total'] = df_ddjj['Bienes_Inmuebles_Valor'] + df_ddjj['Bienes_Muebles_Valor'] + df_ddjj['Activos_Financieros_Valor'] - df_ddjj['Pasivos_Valor']


#5. Calcular 'Patrimonio_2024' y 'Patrimonio_2023' (Si es necesario, basado en logic)
df_ddjj['Patrimonio_2024'] = df_ddjj ['Patrimonio_Total']
df_ddjj['Patrimonio_2023'] = np.where(df_ddjj['Año_Declaracion'] == 2023, df_ddjj['Patrimonio_Total'], 0)

#Si Patrimonio 2023 se calcula a partir de Patrimonio 2024
df_ddjj['Patrimonio_2023'] = np.where(df_ddjj['Año_Declaracion'] == 2024, np.round(df_ddjj['Patrimonio_2024'] * 0.83, 2), df_ddjj['Patrimonio_2023'])
#df_ddjj['Patrimonio_2023'] = df_ddjj['Patrimonio_2024'] * 0.83

# 6. Calcular 'Variacion_Patrimonial_Porcentaje'
df_ddjj['Variacion_Patrimonial_Porcentaje'] = np.where(df_ddjj['Patrimonio_2023'] != 0,
                                                    np.round((df_ddjj['Patrimonio_2024'] - df_ddjj['Patrimonio_2023']) / df_ddjj['Patrimonio_2023'], 3),
                                                    0)

# 7.  Calcular si es sospechoso
df_ddjj['Es_Sospechoso'] = np.where(df_ddjj['Variacion_Patrimonial_Porcentaje'] > 0.15, 1, 0)

df_ddjj = df_ddjj.sort_values('ID_Funcionario')

nombre_archivo_csv = 'df_ddjj.csv'
df_ddjj.to_csv(nombre_archivo_csv, index=False)

print(f"El DataFrame se ha guardado exitosamente en el archivo '{nombre_archivo_csv}'")
df_ddjj
```
